```python
%cd source_code

from glob import glob
import os
import cv2
import matplotlib.pyplot as plt

list_category       = ['half_dress', 'kurta', 'lehenga', 'maxi_dress', 'saree', 'Knee_length_skirt', 'baggy_pant', 'floor_length_skirt', 'formal_pant', 'jeans_pant', 'jeans_short', 'mini_skirt', 'normal_shorts', 'track_pant', 'blouse', 'bomber_jacket', 'coat', 'down_jacket', 'full_cami_tops', 'cami_tops', 'full_shirt', 'sleeved_crop_tops', 'tank_tops', 'full_tshirt', 'tube_tops', 'half_cami_tops', 'half_jacket', 'half_shirt', 'half_tank_tops', 'half_tshirt', 'half_tube_tops', 'hoodie', 'jeans_jacket', 'leather_jacket', 'sleeveless_crop_tops', 'tunic_tops']

total_ads   = len([i for i in glob('../dataset/*/*/*.png') if 'original' not in i])

try:
    with open('logs.txt', 'r') as f:
        log = f.readline()
except:
    log = '0 0 0 0'
    print('logs empty')
starting = log.split(' ')
total_prev = int(starting[-1])
print(starting)
total_count = 0
for cat_num, category in enumerate(list_category):
    list_video_ads = glob(f'../dataset/{category}/*')
    category_size = len(list_video_ads)
    for vid_count, video_ads in enumerate(list_video_ads):
        
        list_ads = glob(f'{video_ads}/*')
        
        number_ads = len(list_ads) - 1
        original_img = cv2.imread(video_ads+'/original.png')
        
        count = 0
        while (count<len(list_ads)):
            ads_count = count
            count+=1
            ads = list_ads[ads_count]

            
            if 'original.png' in ads:
                continue
                
            else:
                if total_prev>=total_count:
                    total_count += 1
                    continue
                    
                per = "{:.2f}".format((total_count/total_ads)*100)
                title = f"{total_count} / {total_ads}    |  Percentage==>{per} %"
                img_ads = cv2.imread(ads)
                
                
                convat = cv2.hconcat([original_img, img_ads])
                cv2.imshow(title, cv2.resize(convat, (600, 700)))
                c = cv2.waitKey(0)
                if  c==115: #'s' 'similar'
                    with open('dataset_sim.txt', 'a') as f:
                        original_path = f"1|{'/'.join(video_ads.split('/')[1:])}+/original.png|{'/'.join(ads.split('/')[1:])}\n"
                        f.write(original_path)
                
                elif c==100:#'d' dissimilar
                    with open('dataset_sim.txt', 'a') as f:
                        original_path = f"0|{'/'.join(video_ads.split('/')[1:])}+/original.png|{'/'.join(ads.split('/')[1:])}\n"
                        f.write(original_path)
                        
                elif c==97:#'a'
                    print('a')
                    count -= 2
                    total_count -= 1
                    with open('dataset_sim.txt', 'r') as f:
                        read_file = f.readlines()
                        
                    with open('dataset_sim.txt', 'w') as f:
                        f.write(''.join(read_file[:-1]))
                        
                    cv2.destroyAllWindows()
                    continue
                    
                elif c==113:#'q'
                    cv2.destroyAllWindows()
                    assert(False)
                    break
                    
                with open('logs.txt', 'w') as f:
                    f.write(f"{cat_num} {vid_count} {ads_count} {total_count}")
                    
                total_count += 1
                cv2.destroyAllWindows()
            
        category_size = category_size - 1  
```
